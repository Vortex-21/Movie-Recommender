```python
data_frame = data_frame[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
data_frame.dropna(inplace = True)
```

```python
data_frame.duplicated().sum()
```

```python
lst=data_frame.iloc[0]['genres']#is a list of dictionaries but in the form of a string
lst
```

```python
import ast
def convert(string):
    L=[]
    for dictionary in (ast.literal_eval(string)):
        L.append(dictionary['name'])
    return L
```

```python
data_frame['genres']=data_frame['genres'].apply(convert)
```

```python
data_frame.head()
```

```python
data_frame['keywords']=data_frame['keywords'].apply(convert)
data_frame.head()
```

```python
data_frame.iloc[0]['cast']
```

```python
def convert_top_3(string):
    L = []
    count=0
    for dictionary in ast.literal_eval(string):
            if(count<3):
                L.append(dictionary['name'])
                count+=1
            else:
                break
    return L
        
```

```python
data_frame['cast']=data_frame['cast'].apply(convert_top_3)
```

```python
data_frame.head()
```

```python
lst = ast.literal_eval(data_frame['crew'][0])
for i in lst:
    if(i['job'] == 'Director'):
        print(i['name'])
```

```python
def get_director(string):
    lst = ast.literal_eval(string)
    L=[]
    for dictionary in lst:
        if(dictionary['job'] == 'Director'):
                L.append(dictionary['name'])
    return L
```

```python
data_frame['crew']=data_frame['crew'].apply(get_director)
data_frame.head()
```

```python
data_frame.head()
```

```python
data_frame['overview'] = data_frame['overview'].apply(lambda x:x.split())
```

```python
data_frame.head()
```

Since the first names of actors can be the same ( in more formal terms because of the presence of duplicacy) we will need to apply a transformation on the words which have spaces the transformation being removal and create a single word

```python
data_frame['genres']=data_frame['genres'].apply(lambda x:[i.replace(" ","") for i in x])#eg:=>Science Fiction becomes ScienceFiction so that we can use it as a single unique entity/tag
data_frame['keywords']=data_frame['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
data_frame['cast']=data_frame['cast'].apply(lambda x:[i.replace(" ","") for i in x])
data_frame['crew']=data_frame['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data_frame.head()
```

```python
data_frame['tags']=data_frame['overview']+data_frame['genres']+data_frame['keywords']+data_frame['cast'] + data_frame['crew']
```

```python
data_frame.head()
```

```python
new_df = data_frame[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    L = []
    for i in text.split():
        L.append(ps.stem(i))
    return " ".join(L)
```

```python
stem('ScienceFiction Playing video games')
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')

```

```python
movie_vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
new_df.head()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sims = cosine_similarity(movie_vectors)#similarity matrix
```

```python
sims.shape
```

```python
sims[0]
```

```python
def recommend(movie):
    movie_idx = new_df[new_df['title']==movie].index[0]
     
    recommended_movies=sorted(list(enumerate(sims[movie_idx])),reverse = True,key = lambda x:x[1])[1:6]
    for i in recommended_movies:
        print(new_df.iloc[i[0]]['title'])
    return recommended_movies
    
```

```python
recommendations = recommend('The Terminator')
```
